```r
annotate_img <- function(img, x_center, y_center, width, height) {
    annotation_raster(
      img,
      xmin=x_center-width/2,
      xmax=x_center+width/2,
      ymin=y_center-height/2,
      ymax=y_center+height/2)
}

plot_UsedCarTrain <- function(x_coords, y_coords, icon_size = c(10, 10)) {
  mapply(
    annotate_icon,
   UsedCarTrain$price,
    x_coords,
    y_coords,
    icon_size[1],
    icon_size[2]
    )
}
plot_UsedCarTrain
```
